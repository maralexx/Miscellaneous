```python
def pairs(arr, n):
    count = 0
    for i in range(n-1):
        for j in range(n-1):
            if arr[i,j]==1 and arr[i+1, j]==1:
                count+=1
            if arr[i,j]==1 and arr[i, j+1]==1:
                count+=1
    for i in range(n-1):
        if arr[i,n-1]==1 and arr[i+1, n-1]==1:
            count+=1
    for j in range(n-1):
        if arr[n-1,j]==1 and arr[n-1,j+1]==1:
            count+=1
    return count
```

```python
def ans(n):
    np.random.seed(0)
    anss = [pairs(bernoulli.rvs(0.1, size=(100,100)), 100) for i in range(n)]
    return sum(anss)/n
```

```python
from collections import defaultdict 

class Graph: 

    def __init__(self,vertices): 
        self.V = vertices
        self.graph = defaultdict(list)
        self.k = []


    def addEdge(self,v,w): 
        self.graph[v].append(w)
        self.graph[w].append(v) 

    def isCyclicUtil(self,v,visited,parent,k):
        visited[v] = True

        for i in self.graph[v]:
            if visited[i]==False : 
                if(self.isCyclicUtil(i,visited,v,k+1)):
                    return True
            elif parent!=i:
                self.k.append(k+1)
                return True
        
        return False
    

    def isCyclic(self):
        j = 0
        visited =[False]*(self.V) 
        for i in range(self.V):
            if len(self.graph[i])>2:
                with open('output.txt', 'w') as fout:
                    fout.write('NO')
                return None
            if visited[i] == False: 
                if self.isCyclicUtil(i,visited,-1,0)== True:
                    if self.k[j]%2==1:
                        with open('output.txt', 'w') as fout:
                            fout.write('NO')
                        return None
                    else:
                        j+=1
        with open('output.txt', 'w') as fout:
            fout.write('YES')
        return None

with open('input.txt', 'r') as fin:
    for i, line in enumerate(fin):
        if i==0:
            g = Graph(int(line.split(' ')[0])+1)
        else:
            lst = list(map(int, line.split(' ')))
            g.addEdge(lst[0], lst[1])
g.isCyclic()
```

```python
using namespace std;

int main() {
    int l = 1, r = 1000000;
    while (l != r) {
        int mid = (l + r + 1) / 2;
        printf("%d\n", mid);
        fflush(stdout);

        char response[3];
        scanf("%s", response);
        if (strcmp(response, "<") == 0)
            r = mid - 1;
        else
            l = mid;
    }

    printf("! %d\n", l);
    fflush(stdout);
}
```
